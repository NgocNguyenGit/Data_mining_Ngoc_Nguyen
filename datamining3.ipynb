```python
#EX3
#Initialisation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("data_cluster.csv")

np.random.seed(200)

k = 3

#centroids[i] = [x, y]

centroids = {
    i+1: [np.random.randint(24, 50), np.random.randint(30, 59)]
    for i in range(k)
}

fig = plt.figure(figsize=(5,5))
plt.scatter(df['X'],df['Y'],color='k')
colmap = {1:'r', 2:'g',3:'b',4:'y'}
for i in centroids.keys():
    plt.scatter(*centroids[i], color = colmap[i])

plt.xlim(24,50)
plt.ylim(30, 59)
plt.show()
```

```python
#Assignment stage

def assignment(df, centroids):
    for i in centroids.keys():
        #sqrt((x1-x2)^2 + (y1-y2)^2)
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df['X'] - centroids[i][0])**2 
                + (df['Y'] - centroids[i][1])**2
            )
        )
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis = 1)    
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    df['color'] = df['closest'].map(lambda x: colmap[x])
    return df

df = assignment(df, centroids)
print(df.head())

fig = plt.figure(figsize=(5,5))
plt.scatter(df['X'],df['Y'],color = df['color'], alpha = 0.5, edgecolor = 'k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color = colmap[i])

plt.xlim(24,50)
plt.ylim(30, 59)
plt.show()
    
```

```python
#Update stage

import copy

old_centroids = copy.deepcopy(centroids)

def update(k):
    for i in centroids.keys():
        centroids[i][0] = np.mean(df[df['closest'] == i]['X'])
        centroids[i][1] = np.mean(df[df['closest'] == i]['Y'])
    return k

centroids = update(centroids)

fig = plt.figure(figsize = (5,5))
ax = plt.axes()


fig = plt.figure(figsize=(5,5))
plt.scatter(df['X'],df['Y'],color = df['color'], alpha = 0.5, edgecolor = 'k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color = colmap[i])

plt.xlim(24,50)
plt.ylim(30, 59)

for i in old_centroids.keys():
    old_x = old_centroids[i][0]    
    old_y = old_centroids[i][1]
    dx = (centroids[i][0] - old_centroids[i][0]) * 0.75
    dy = (centroids[i][1] - old_centroids[i][1]) * 0.75
    ax.arrow(old_x, old_y, dx, dy, head_width = 2, head_length = 3, fc = colmap[i], ec = colmap[i])
    
plt.show()
```

```python
#Repeat assignment stage

df = assignment(df, centroids)

#Plot results
fig = plt.figure(figsize=(5,5))
plt.scatter(df['X'],df['Y'],color='k')
colmap = {1:'r', 2:'g',3:'b',4:'y'}
for i in centroids.keys():
    plt.scatter(*centroids[i], color = colmap[i])

plt.xlim(24,50)
plt.ylim(30, 59)
plt.show()
```

```python
#Continue until all assigned categories don't change anymore

while True:
    closest_centroids = df['closest'].copy(deep = True)
    centroids = update(centroids)
    df = assignment(df, centroids)
    if closest_centroids.equals(df['closest']):
        break

fig = plt.figure(figsize=(5,5))
plt.scatter(df['X'],df['Y'], color = df['color'], alpha = 0.5, edgecolor = 'k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color = colmap[i])

plt.xlim(24,50)
plt.ylim(30, 59)

plt.show()
```

```python
df = pd.read_csv('data_cluster.csv')
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 3)
kmeans.fit(df)
```

```python
labels = kmeans.predict(df)
centroids = kmeans.cluster_centers_

fig = plt.figure(figsize=(5,5))
colors = map(lambda x: colmap[x + 1], labels)
colors1 = list(colors)
plt.scatter(df['X'],df['Y'], color = colors1, alpha = 0.5, edgecolor = 'k')
for idx, centroids in enumerate(centroids):
    plt.scatter(*centroids, color = colmap[idx + 1])

plt.xlim(24,50)
plt.ylim(30,59)

plt.show()
```

```python
#EX5
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import collections

#load data
data = pd.read_csv("data_cluster.csv")
data = np.matrix(data)

num = np.arange(0, np.size(data, 0))
num = np.transpose(np.matrix(num))
data_with_cluster = np.concatenate((data, num), axis = 1)


def flatten(x):
    if isinstance(x, collections.Iterable):
        return [a for i in x for a in flatten(i)]
    else:
        return [x]
```

```python
def runKH(data, graph):
    closest = []
    distance = []
    for i in range(np.size(data, 0)):
        temp = np.linalg.norm(data[i, :2] - data[:, :2], None, 1)
        closest += [np.argpartition(temp, 2)[1]]
        distance += [temp[np.argpartition(temp, 2)[1]]]
    
    idx1 = np.argmin(distance)
    idx2 = closest[idx1]
    new_cluster = np.mean((data[idx1,:2], data[idx2,:2]), 0)
    new_cluster = np.concatenate((new_cluster, [[data[idx1, 2]]]), axis = 1)

    graph[int(data[idx1, 2])] = [graph[int(data[idx1, 2])], graph[int(data[idx2, 2])]]
    data = np.delete(data, [idx1, idx2], 0)    
    data = np.append(data, new_cluster, axis = 0)

    return (data, graph)
```

```python
graph = {}
for i in range(40):
    graph[i] =[i]
```

```python
K = 3
for i in range(np.size(data, 0) - K):
    (data_with_cluster, graph) = runKH(data_with_cluster, graph)
print(data_with_cluster)
```

```python
l = {}
for i in range(np.size(data_with_cluster, 0)):
    l[i] = flatten(graph[data_with_cluster[i, 2]])
    plt.scatter(np.array(data[:,0][l[i]]), np.array(data[:,1][l[i]]), 10, c = np.random.rand(3,)) 
```
