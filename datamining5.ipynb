```python
import pandas as pd
import numpy as np

data = pd.read_csv('olympic.csv')
print(data)
```

```python
#1: Create scatter plots for 3 pairs of variables.
import matplotlib.pyplot as plt

x = data['Name']
y = data['run800m']

x = data['Name']
b = data['longjump']

x = data['Name']
e = data['shot']

plt.scatter(x, y, color = 'r') #pair 1
plt.scatter(x, b, color = 'g') #pair 2
plt.scatter(x, e, color = 'b') #pair 3

plt.xlabel('Name')
plt.ylabel('Grades Scored')

plt.show()
```

```python
#2: Find a correlation matrix
data = np.array(data)
data = data[:, 1:]
data = data.astype("float")
data_c = np.cov(data)
print(data_c)
```

```python
(U, S, V) = np.linalg.svd(data_c) #U eigenvector, eigenvalue, V nonsense
print(U)
print(S)
```
